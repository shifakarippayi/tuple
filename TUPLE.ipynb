The union () method returns a new set with all itesms from both sets

```python
names={'shifa','shree','pranav','abishak'}
names1={'shifa','shree','pranav','abishek','ajay'}

names3=set.union(names,names1)
print(names3)
```

```python
names={'ajay','shifa','aswathy','pranav','abishek'}
names1={'shifa','aswathy','pranav','abishek'}

names3=set.difference(names1,names)
print(names3)
```
